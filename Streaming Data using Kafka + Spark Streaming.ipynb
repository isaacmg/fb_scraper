# Streaming Data Using Kafka
Welcome to our streaming data notebook. The goal of this notebook is to use Kafka to stream the FB Posts in realtime as they are scraped into Spark Streaming for analysis. 

```python
from kafka import KafkaConsumer
import avro.schema
import avro.io
import io
import json
consumer = KafkaConsumer('fb',bootstrap_servers=['localhost:9092'])
print("begin")

schema_path="fb_scheam.avsc"
schema1 = avro.schema.parse(open(schema_path).read())

for msg in consumer:
    print("getting mesage")
    bytes_reader = io.BytesIO(msg.value)
    decoder = avro.io.BinaryDecoder(bytes_reader)
    reader = avro.io.DatumReader(schema1)
    message = reader.read(decoder)
    print(message)
```

```python
#Now that the test is working let's try spark streaming 
sc.stop()
ssc.stop()
import findspark
findspark.init()
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
import findspark
import sys
sc = SparkContext(appName="PythonStreamingDirectKafkaWordCount")
ssc = StreamingContext(sc, 2)
kvs = KafkaUtils.createStream(ssc, "localhost:9092", "name", {"test":1})
lines = kvs.map(lambda x: x[1])
counts = lines.flatMap(lambda line: line.split(" ")) \
        .map(lambda word: (word, 1)) \
        .reduceByKey(lambda a, b: a+b)
counts.pprint()
print(counts)
ssc.start()
ssc.awaitTermination()
ssc.stop()

##UGHHHHHHHHHHHHHHHHHHHHH SPARK IS NOW THROWING ERRORS!!!!!!!!!
```
